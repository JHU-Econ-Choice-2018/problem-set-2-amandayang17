```python
# Some initial setup

import sys 
import os
sys.path.insert(0, os.path.abspath('../lib'))


from matplotlib import pyplot as plt
import numpy as np
plt.style.use('seaborn-darkgrid')
palette = plt.get_cmap('Dark2')
from copy  import deepcopy

from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
```

```python
Epsilon=0.33
DiscFac=0.96
PopGrowth=1.01
YearsPerGeneration=30
kMax=0.05
#Initialk=0

# Define some parameters
Beta = DiscFac**YearsPerGeneration
xi = PopGrowth**YearsPerGeneration
Q = (1-Epsilon)*(Beta/(1+Beta))/xi
kBar = Q**(1/(1-Epsilon))
    
# Plot the OLG capital accumulation curve and 45 deg line
plt.figure(figsize=(9,6))
kt_range = np.linspace(0, kMax, 1000)

# calculate k_t+1 and plot the OLG capital accumulation curve and 45 deg line
ktp1 = Q*kt_range**Epsilon
plt.plot(kt_range, ktp1, 'b-', label='Capital accumulation curve')
plt.plot(kt_range, kt_range, 'k--', label='45 Degree line')

plt.legend()
plt.xlim(0 ,kMax)
plt.ylim(0, kMax)
plt.xlabel('$k_t$')
plt.ylabel('$k_{t+1}$')
plt.plot(kBar, kBar, 'ro', label='kBar')
plt.axvline(x=kBar,ymin=0,ymax=1,color='b',linestyle='--')
plt.text(0.038, -0.003, r'$\bar{k}_0$',color='r',fontsize=14)
plt.show()
```

```python
Epsilon=0.33
DiscFac=0.96
PopGrowth_1=1.01
PopGrowth_2=1.03
YearsPerGeneration=30
kMax=0.05
#Initialk=0

# Define some parameters
Beta = DiscFac**YearsPerGeneration
xi_1 = PopGrowth_1**YearsPerGeneration
xi_2 = PopGrowth_2**YearsPerGeneration
Q_1 = (1-Epsilon)*(Beta/(1+Beta))/xi_1
Q_2 = (1-Epsilon)*(Beta/(1+Beta))/xi_2
kBar_1 = Q_1**(1/(1-Epsilon))
kBar_2 = Q_2**(1/(1-Epsilon))

    
# Plot the OLG capital accumulation curve and 45 deg line
plt.figure(figsize=(10,7))
kt_range = np.linspace(0, kMax, 1000)

# calculate k_t+1 and plot the OLG capital accumulation curve and 45 deg line
ktp1_1 = Q_1*kt_range**Epsilon
ktp1_2 = Q_2*kt_range**Epsilon
plt.plot(kt_range, ktp1_1, 'b-', label='Capital accum curve for 0.01 PopGrowth')
plt.plot(kt_range, ktp1_2, 'g-', label='Capital accum curve for 0.03 PopGrowth')
plt.plot(kt_range, kt_range, 'k--', label='45 Degree line')

plt.legend()
plt.xlim(0 ,kMax)
plt.ylim(0, kMax)
plt.xlabel('$k_t$')
plt.ylabel('$k_{t+1}$')
plt.plot(kBar_1, kBar_1, 'ro', label='kBar_1')
plt.plot(kBar_2, kBar_2, 'ro', label='kBar_2')
plt.axvline(x=kBar_1,ymin=0,ymax=1,color='b',linestyle='--')
plt.axvline(x=kBar_2,ymin=0,ymax=1,color='g',linestyle='--')
plt.text(0.038, -0.003, r'$\bar{k}_0$',color='r',fontsize=14)
plt.text(0.0155, -0.003, r'$\bar{k}_1$',color='r',fontsize=14)
plt.show()
```

```python
Epsilon=0.33
DiscFac=0.96
PopGrowth_1=1.01
PopGrowth_2=1.03
YearsPerGeneration=30
kMax=0.05
Initialk=0

# Define some parameters
Beta = DiscFac**YearsPerGeneration
xi_1 = PopGrowth_1**YearsPerGeneration
xi_2 = PopGrowth_2**YearsPerGeneration
Q_1 = (1-Epsilon)*(Beta/(1+Beta))/xi_1
Q_2 = (1-Epsilon)*(Beta/(1+Beta))/xi_2
kBar_1 = Q_1**(1/(1-Epsilon))
kBar_2 = Q_2**(1/(1-Epsilon))

    
# Plot the OLG capital accumulation curve and 45 deg line
plt.figure(figsize=(13,9))
kt_range = np.linspace(0, kMax, 1000)

# calculate k_t+1 and plot the OLG capital accumulation curve and 45 deg line
ktp1_1 = Q_1*kt_range**Epsilon
ktp1_2 = Q_2*kt_range**Epsilon
plt.plot(kt_range, ktp1_1, 'b-', label='Capital accum curve for 0.01 PopGrowth')
plt.plot(kt_range, ktp1_2, 'g-', label='Capital accum curve for 0.03 PopGrowth')
plt.plot(kt_range, kt_range, 'k--', label='45 Degree line')

plt.xlim(0 ,kMax)
plt.ylim(0, kMax)
plt.xlabel('$k_t$')
plt.ylabel('$k_{t+1}$')
plt.plot(kBar_1, kBar_1, 'ro', label=r'$\bar{k}_0$')
plt.plot(kBar_2, kBar_2, 'bo', label=r'$\bar{k}_1$')
plt.axvline(x=kBar_1,ymin=0,ymax=1,color='b',linestyle='--')
plt.axvline(x=kBar_2,ymin=0,ymax=1,color='g',linestyle='--')
#plt.text(0.038, -0.003, r'$\bar{k}_0$',color='r',fontsize=14)
#plt.text(0.0155, -0.003, r'$\bar{k}_1$',color='r',fontsize=14)
plt.legend()

# plot the path
Initialk=kBar_1
kt_ar = Initialk
ktp1_ar = kt_ar
for i in range(3):
    plt.arrow(kt_ar, ktp1_ar, 0., Q_2*kt_ar**Epsilon-ktp1_ar,
                length_includes_head=True,
                lw=0.01,
                width=0.0003,
                color='black',
                edgecolor=None)
    plt.arrow(kt_ar, Q_2*kt_ar**Epsilon, Q_2*kt_ar**Epsilon-kt_ar, 0.,
                length_includes_head=True,
                lw=0.01,
                width=0.0003,
                color='black',
                edgecolor=None)
    kt_ar = Q_2*kt_ar**Epsilon
    ktp1_ar = kt_ar
    i = i+1  

plt.show()
```
